# Broadcasting
- Treatment of arrays of different shape during arithmetic operations.
- Usual arithmetic operations are arried out between corresponding elements of two different arrays of sam shape.<br>

2 arrays are broadcastable if:
- They are both of the same type
- Arrays have same number of dimensions,  length of each dimension is either a common length or 1
- If the dimensions are not the same then the array obtained by prepending the shape by 1 must satisfy the above property.

```python
x = np.array([3,1,2,5])
y = x[::-1]
print(5*x,'\n')
print(y*x,'\n')

x = np.array([[1,2,3,4],[4,5,6,7]])
y = np.arange(1,5)
print(x*y,'\n')
```

# Arithmetic operations
All operations are caried out element wise

```python
x = np.arange(5)
y = np.arange(5,10)
print('x = ',x)
print('x - y = ',x-y)
print('x*y =',x*y)
print('x/y = ',x/y)
print('y//x = ',y//x)
print('x**2 = ',x**2)
print('x**y = ',x**y)
print('x<2 = ',x<2)
print('x<y = ',x<y)
print('x+1 = ',x+1)
```

# Reshaping arrays:
The following commands return an array with modified shape.
- arr.ravel() returns a vector
- arr.reshape() chages shape according to parameters **Where -1 is used as parameter, the parameter is evaluateed by the program**
- arr.T transpose
- arr.swapaxes(axis_1,axis_2)

```python
x = np.array([[[1,2,3,1]
             ,[3,0,3,2]
             ,[1,1,1,3]],
             [[1,0,1,3]
             ,[4,5,6,2]
             ,[0,0,0,1]]])
print(x.shape)
print(x)
print('\n',x.swapaxes(0,1))
print('\n',x.swapaxes(1,2))
print('\n',x.swapaxes(0,2))
```
