# Subtle points realisded during implimentation of linear regression:
- convergence: error converges to a value
- learning rate: should change with magnitude of input parameters input parameters

```python
import matplotlib.pyplot as plt
import copy
import math
```

```python
x = [2104,1600,2400,1416,3000,1985,1534,1427,1380,1494]
y = [400,330,369,232,540,300,315,199,212,242.5]
#x = [0,1,2,3]
#y = [-1,1,2,3]
t = copy.copy(x)
x.sort()
c = []
for i in range(0,len(y)):
    c.append(y[t.index(x[i])])
y = c
```

```python
c0 = 0
c1 = 0
a = 5
thr = 1e-2
err = 0
for i in y:
    err = err + i*i
err1 = 0
```

```python
#for i in range(0,300):
while abs(err1-err)>thr:
    t1 = 0
    t0 = 0
    for i in range(len(x)):
        t0 = t0 + (c0*x[i]+c1-y[i])*x[i]
        t1 = t1 + (c0*x[i]+c1-y[i])*1
    c0 = c0 -a*t0/abs(t0+t1)
    c1 = c1 -a*t1/abs(t0+t1)
    err = err1
    err1 = 0
    for i in range(len(x)):
        err1 = err1+(c0*x[i]+c1-y[i])*(c0*x[i]+c1-y[i])
    if err1>err:
        a = a/10
    print([c0,c1])
```

```python
h = []
for i in x:
    h.append(c0*i+c1)
plt.plot(x,h)
plt.plot(x,y,'o')
print(h)
print(y)
print(x)
print('rms error:',math.sqrt(err1/len(x)))
```
