# global keyword
*global var_name* means that the variable identifier var_name refers to the global variable in this block.<br>
the variable **cannot** be used before the global declation in the block<br>
**an identifier can either represent a global variable or a local variable throughout the program. <br> Hence the global declaration cant be used in the middle of a block if the variable has been defined previouly locally. Similarly if a variable has been defined to be global, it cannot be declared later or before in the block locally.**

```python
def shittyFunc():
    print('in func shittyFunc')
    global var 
    print(var)
    var = var + var
    print(var)
    print('exitFunc')
    
var = 'shit '
print(var)
shittyFunc()
print(var)
    
```

# References - Mighty caution 
<hr>
When you assign a list to a variable, you <br>
are actually assigning a list reference to the variable. A reference is a value <br>
that points to some bit of data, and a list reference is a value that points to a <br>
list. Hence changing anyone in the program will affect the other.
<hr>

 list variables don’t actually contain lists—they contain references
to lists. <br>
But for strings and integer values, variables simply contain the <br>
string or integer value. Python uses references whenever variables must <Br>
store values of **mutable data types, such as lists or dictionaries**. For values <br>
of **immutable data types such as strings, integers, or tuples**, Python variables will store the value itself.<br>
<hr>
    
Lists and dictionaries are passed bt reference to functions. Hence the changes made to parameter lists and dictionaries<br>
affect the actual list/dictionary. <br> Immutable data types or ints and floats are copied to the function and hence <br>
cannot / willnot  be affected<br>

```python
list1 = ['lol','i','suck']
list2 = list1
list1.append('dix')
print(list1)
print(list2)
```

# Copy module and *copy* and *deep copy* functions
Suppose you dont want to assign a copy of a list inside of creating another reference to the same list.<br>
You can use the copy.copy(list_name) function to create a copy of a list and a copy.deepcopy(list_name) to create <br>
a copy of inner lists as well list within a list.

**deep copy creates the depest copies and any level of inception among lists is copied and not referenced**

```python
import copy
l1 = [1,2,3,4]
l2 = copy.copy(l1)
l1.append(5)
print(l1,'\n',l2)
```

```python
import copy
l1 = [1,2,3,[1,2,3]]
l2 = copy.copy(l1)
l1.append(4)
l1[3].append(4) # here the inner list has not been copied and hence modifying it changes both the list
print(l1)
print(l2)
```

```python

import copy
l1 = [1,2,3,[1,2,3,[4,1]]]
l2 = copy.deepcopy(l1)
l1.append(4)
l1[3].append(4) 
l1[3][3].append(7)
print(l1)
print(l2)
```

```python
# PROGRAM TO COUNT THE OCCURENCE OF EACH CHARACTER IN A STRING 
message = 'This shit is ice cold, michelle phiper that white gold. This one for them hooded girls dem good girls straight masterpiece'
occurence = {}

for char in message:
    occurence.setdefault(char,0)
    occurence[char] +=1
    
while occurence != {}:
    arr = list(occurence.keys())
    M = arr[0]
    for i in arr:
        if(occurence[i]>occurence[M]):
            M = i
    print(M,':',occurence[M])
    del occurence[M]
```
