f-strings

```python
# .format

# use {}  as place holders
print("This is a string {}".format("_insrerted_text_"))

# use indexing for  multiple insertions
print("\nThis is {} {} {}".format('an','cold','ice'))
print("This is {0} {2} {1}".format('an','cold','ice'))
print("This is {1} {1} {1}".format('an','cold','ice'))

# use assignment key values
print("\nThis is {a} {i} {c}".format(a='an',i='ice',c='cold'))
print("This is an {i} {i} {i}".format(a='an',i='ice',c='cold'))
```
